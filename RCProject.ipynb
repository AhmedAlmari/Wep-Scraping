```python
def webscraping(urls, dictionary):
    output = []
    image_urls = dictionary.get("image_URL",[])
    for i, url in enumerate(urls):
        values = {}
        for column,value in dictionary.items():
            if value is None:
                values[column] = None
            elif column == "watch_URL":
                values[column] = url
            elif column == "image_URL":
                if i < len(image_urls):
                    values[column]= image_urls[i]
                else:
                    values[column] = None
            else:
                xpath = value if value.startswith("/") else None
                if xpath:
                    response = requests.get(url)
                    soup = BeautifulSoup(response.content,'html.parser')
                    dom = etree.HTML(str(soup))
                    result = dom.xpath(xpath)
                    if result:
                        values[column] = result[0].strip() if isinstance(result[0], str) else result[0].text
                    else:
                        values[column] = None
                else:
                    values[column] = value
        output.append(values)
    return output
```
